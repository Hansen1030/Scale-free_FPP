```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python

def bezier_curve(start, end, cost, ax):
    direction=1
    if start>end:
        direction=-1
    p0 = np.array([start, 0])
    p1 = np.array([(start + end) / 2, (direction)*cost]) # height of the control point
    p2 = np.array([end, 0])

    t = np.linspace(0, 1, num=100)[:, None]
    curve = (1 - t)**2 * p0 + 2 * (1 - t) * t * p1 + t**2 * p2
    ax.plot(curve[:, 0], curve[:, 1], '-')
def plot(data):
    fig, ax = plt.subplots(figsize=(12, 8))
    for index, row in data.iterrows():
        bezier_curve(row['start'], row['end'], row['cost'], ax)
    ax.set_xlabel('Node position')
    ax.set_ylabel('cost')
    ax.set_title('shortest path')

    plt.show()
```

```python
path="path/2.200000_0.800000_20000"
temp_data=np.loadtxt(path)
print(temp_data[0])
data_list=[]
data_frame=[]
single=[]
n=int(path.split("_")[2])
start=n*0.2
end=n*0.8
for i in temp_data:
    i[0]=(i[0]-start)/(end-start)
    i[1]=(i[1]-start)/(end-start)
    single.append(i)
    if i[1]==1:
        data_list.append(single)
        data_frame.append(pd.DataFrame(single,columns=["start","end","cost"]))
        single=[]

data_frame[0]
```

```python
for i in data_frame:
    plot(i)
data_frame[1]
```
