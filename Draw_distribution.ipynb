```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def scale_function(dataset):
    mean=dataset.mean()
    var=dataset.var()
    return (dataset-mean)/np.sqrt(var)
def draw_distribution(dataset,names,cdf=True,shade=False):
    plt.figure(figsize=(8, 4))
    for i in range(len(dataset)):
        dataset[i]=scale_function(dataset[i])
        sns.kdeplot(dataset[i], label=names[i], cumulative=cdf,fill=False)
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.title('Probability Density Functions of Both Datasets')
    plt.legend()
    plt.show()
```

```python
dataset1 = np.loadtxt('data_new/3.000000_0.800000_10000')
dataset2= np.loadtxt('data_new/3.000000_0.800000_20000')
data=np.vstack((dataset1, dataset2))
```

```python
names=['alpha=3, gamma=0.8,N=10000',
       'alpha=3, gamma=0.8,N=20000']
draw_distribution(data,names,cdf=True)
```
