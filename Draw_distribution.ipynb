```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def scale_function(dataset):
    mean=dataset.mean()
    var=dataset.var()
    return (dataset-mean)/np.sqrt(var)
def draw_distribution(dataset,names,cdf=True,shade=False):
    plt.figure(figsize=(8, 4))
    for i in range(len(dataset)):
        dataset[i]=scale_function(dataset[i])
        name=names[i].split("_")
        sns.kdeplot(dataset[i], label="alpha="+name[0]+", gamma="+name[1]+"'N="+name[2], cumulative=cdf,fill=False)
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.title('Probability Density Functions of Both Datasets')
    plt.legend()
    plt.show()
    pass
def data_generate(names):
    data=np.loadtxt('data_new/'+names[0])
    for i in range(len(names)):
        if i==0 : continue
        data=np.vstack((data,np.loadtxt('data_new/'+names[i])))
    return data
```

```python
# The name is defult setting your data is under the folder 'data_new/' if not, remmenber put you data in 'data_new/' folder
names=np.array(['3.000000_0.800000_10000','3.000000_0.800000_20000'])
data=data_generate(names)
```

```python
draw_distribution(data,names,cdf=True)
```
