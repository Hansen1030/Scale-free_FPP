```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def data_generate(names):
    data=np.loadtxt('data_new/'+names[0])
    for i in range(len(names)):
        if i==0 : continue
        data=np.vstack((data,np.loadtxt('data_new/'+names[i])))
    return data

def scale_function(dataset,N,alpha):
    mean=dataset.mean()
    median=np.median(dataset)
    # return N/(pow(dataset,1/(alpha-2.5)))
    return (dataset)/(pow(N,(alpha-2)))
    # return dataset/median

def draw_distribution(dataset,names,cdf=True,shade=False):
    plt.figure(figsize=(10, 6))
    for i in range(len(dataset)):
        name=names[i].split("_")
        dataset[i]=scale_function(dataset[i],int(name[2]),float(name[0]))
        sns.kdeplot(dataset[i], label="alpha="+name[0]+", gamma="+name[1]+",N="+name[2], cumulative=cdf,fill=False)
    plt.xlabel('Value')
    plt.ylabel('Density')
    if cdf:
        plt.title('Cumulative Distribution Functions of Both Datasets')
    else:
        plt.title('Probability Density Functions of Both Datasets')
    plt.legend()
    plt.show()
    pass
```

```python
# The name is defult setting your data is under the folder 'data_new/' if not, remmenber put you data in 'data_new/' folder
names=np.array(['3.000000_0.800000_10000','3.000000_0.800000_10000_fake'])
data=data_generate(names)
draw_distribution(data,names,cdf=True)
```
