```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('2.5 3 5000',header = None)
df.columns = ['Time']
df
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(20, 12))
sns.histplot(data=df, x=df['Time'])
plt.title('Distribution of alpha = 3.5, gamma = 3, n = 5000')
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(20, 12))
sns.kdeplot(data=df, x=df['Time'])
plt.title('Distribution of alpha = 3.5, gamma = 3, n = 5000')
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(20, 12))
sns.boxplot(data=df, x=df['Time'])
plt.title('Distribution of alpha = 3.5, gamma = 3, n = 5000')
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.show()
```
